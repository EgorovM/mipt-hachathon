```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("Data/train.xlsx")
```

```python
data = data[np.isnan(data["well id"]) != True]
data = data[np.isnan(data["goal"]) != True]

data["well id"] = data["well id"].astype('int')
```

```python
min_depth = 100
min_delta_depth = 300
```

```python
ALL_SENSORS = ['bk', 'GZ1', 'GZ2', 'GZ3', 'GZ4', 'GZ5', 'GZ7', 'DGK', 'NKTD', 'NKTM', 'NKTR', 'ALPS']
sensors_cost = {
    "bk" : 2450,
    "GZ1" : 2050,
    "GZ2" : 2050,
    "GZ3" : 2050,
    "GZ4" : 2050,
    "GZ5" : 2050,
    "GZ7" : 2050,
    "DGK" : 1300,
    "NKTD" : 2050,
    "NKTM" : 2050,
    "NKTR" : 2050,
    "ALPS" : 1150
}


min_depth = int(min(data["depth, m"]))
max_depth = int(max(data["depth, m"])) + 1
```

```python
sensors = ALL_SENSORS
expenses = sum([sensors_cost[sensor] for sensor in sensors])
```

```python
max_cost = 0
min_depth_rec = 0
min_delta_delth_rec = 0

for min_depth in range(100, 2900, 100):
    for delta in range(0, 3000, 100):
        delta_depth = min_delta_depth + delta
        
        df = data[(data["depth, m"] > min_depth) & (data["depth, m"] < min_depth + delta_depth)]
        val = df["goal"].value_counts()
        
        try:
            cost = 4.15 * 0.7 * val[1] * 0.1 * 100 * 860
            print()
        except:
            cost = 0
        
        cost -= expenses * delta_depth
        
        print(f"При глубинах {min_depth} - {min(min_depth + delta_depth, max_depth)} м.")
        print(f"Выручка составляет: %.2f рублей" % cost, "\n")
        
        if cost > max_cost:
            max_cost = cost
            min_depth_rec = min_depth
            min_delta_delth_rec = delta_depth

print("Наилучший показатель")
print(f"При глубинах {min_depth_rec} - {min(min_depth + delta_depth, max_depth)} м.")
print(f"delta: {min_delta_delth_rec}")
print(f"Чистая выручка составляет: ", max_cost, "\n")
        
```
